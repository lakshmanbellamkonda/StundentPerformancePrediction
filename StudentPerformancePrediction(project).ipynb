```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Documents/StudentsPerformance.csv")
print(df.head())
```

```python
#Gender distrubution
plt.figure(figsize = (5,5))
sns.countplot(data = df, x = "gender")
plt.show()
```

```python
gb = df.groupby("parental level of education").agg({"math score":'mean',"reading score":"mean","writing score":"mean"})

print(gb)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb, annot = True)
plt.title("Realationship between parent's eductaion and student's course")
plt.show()
```

```python
gb1 = df.groupby("test preparation course").agg({"math score":'mean',"reading score":"mean","writing score":"mean"})

print(gb1)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb1, annot = True)
plt.show()
```

```python
#from the given chat we have concluded that completing the test preparation course has a positive 
# the impact on student performance.
```

```python
groupA = df.loc[(df["race/ethnicity"] == "group A")].count()
print(groupA)
```

```python
#distribution group 
groupA = df.loc[(df["race/ethnicity"] == "group A")].count()
groupB = df.loc[(df["race/ethnicity"] == "group B")].count()
groupC = df.loc[(df["race/ethnicity"] == "group C")].count()
groupD = df.loc[(df["race/ethnicity"] == "group D")].count()
groupE = df.loc[(df["race/ethnicity"] == "group E")].count()

l = ["group A", " group B", "group C", "group D", "group E"]
mlist = [groupA["race/ethnicity"], groupB["race/ethnicity"], groupC["race/ethnicity"], groupD["race/ethnicity"], groupE["race/ethnicity"]]

print(mlist)
plt.pie(mlist, labels =l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnics Groups")
plt.show()



```
